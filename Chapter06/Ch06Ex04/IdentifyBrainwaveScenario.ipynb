```
from pyspark.sql.functions import lit
scenario = 'ClassicalMusic'
readPath = "wasbs://brainjammer@csharpguitaradeblob.blob.core.windows.net/*.avro"
df = spark.read.format("avro").load(readPath)
df = df[(df['ALPHA'].between(3.8431, 4.3924)) | (df['BETA_H'].between(1.2008, 1.2259)) | \
        (df['BETA_L'].between(1.9012, 1.9363)) |  (df['GAMMA'].between(0.8502, 0.8675)) | \
        (df['THETA'].between(14.6206, 16.8396))]
df.select(lit(scenario).alias("SCENARIO"), "*").show()
```
